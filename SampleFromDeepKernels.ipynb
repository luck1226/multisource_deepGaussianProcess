```python
d = 1 # input dimension
var = 1. # variance
theta = 0.2

k = GPy.kern.RBF(d,var,theta)
```

```python
theta = np.asarray([0.2,0.5,1.,2.,4.])

for t in theta:
    k['.*lengthscale']=t
    k.plot()

pb.legend(theta)
```

```python
k = GPy.kern.RBF(input_dim=1,lengthscale=0.2)

X = np.linspace(0.,1.,500) # 500 points evenly spaced over [0,1]
X = X[:,None] # reshape X to make it n*D

mu = np.zeros((500)) # vector of the means
C = k.K(X,X) # covariance matrix
# Generate 20 sample path with mean mu and covariance C
Z = np.random.multivariate_normal(mu,C,5)
pb.figure() # open new plotting window

for i in range(5):
    pb.plot(X[:],Z[i,:])
```

```python
kRBF = deepRBF(input_dim=1,variance1=2.,lengthscale=0.2,variance2=1)
CRBF = kRBF.K(X,X) # covariance matrix

ZRBF = np.random.multivariate_normal(mu,CRBF,5)
pb.figure() # open new plotting window

for i in range(5):
    pb.plot(X[:],ZRBF[i,:])
```
